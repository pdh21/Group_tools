## Astropy

### Table reading

First, lets use astropy to read in a table:

```python
from astropy.io import fits
hdulist_table=fits.open('mod_cosmos-xid-pepprior-0512.fits')
```

Now, lets see how many extensions there are in the fits file:

```python
print hdulist_table
```

You can see there are two, the primary header, and a table object. Lets see what columns are in the table:

```python
hdulist_table[1].columns
```

Lets plot a histogram of $250 \mathrm{\mu m}$ flux.

```python
%matplotlib inline
#this command tells notebook to keep plots in notebook
import numpy as np
import pylab as plt
plt.figure()
plt.hist(hdulist_table[1].data['F250'],bins=np.arange(0,100,1))
```

### Image Reading
Lets open an image

```python
hdulist_image=fits.open('COSMOS_image_500_SMAP_v4.1.fits')
```

Lets view what extensions are available

```python
print hdulist_image
```

You can see there is the primary header, and three image extensions. Now lets plot the image

```python
plt.figure()
plt.imshow(hdulist_image[1].data)
plt.colorbar()
```

## APLPY
APLPY is a module for making plots with astronomy images. It will be more useful if you want to make images for your projects, with the correct RA and Dec. on the axis, and over plotting catalogues. Lets import the module and make a plot of the same image and catalogue.

```python
import aplpy
```

```python
## ----this is to avoid printing a deprication warning involving matplotlib
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning) 
##--------------------------------



gc=aplpy.FITSFigure(hdulist_image,hdu=1)
gc.show_grayscale()
gc.show_markers(hdulist_table[1].data['RA_J2000'], hdulist_table[1].data['Dec_J2000'], layer='marker_set_1', edgecolor='red',
                facecolor='none', marker='o', s=5, alpha=0.5)


```
